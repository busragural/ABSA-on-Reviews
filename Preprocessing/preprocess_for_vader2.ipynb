```python
import pandas as pd 
import helpers
import re
```

```python
path = "/Users/busragural/Desktop/4.1/ABSA/Data/dataset/processingWithGemini.csv"
df = pd.read_csv(path)
df.head()
```

```python
df = df[df["Split_Review"].notna() & (df["Split_Review"].str.strip() != "")]
```

```python
df['Split_Review'] = df['Split_Review'].apply(helpers.detect_smiley)
df
```

```python
df['Split_Review'] = df['Split_Review'].apply(helpers.replace_emoji)
df['Split_Review']
```

```python
df['Split_Review'] = df['Split_Review'].str.replace(r'\s+', ' ', regex=True)
df['Split_Review']
```

```python
def add_space_after_punctuation(text):
    text = re.sub(r'(?<!\.)\.(?!\.)|(?<!\!)\!(?!\!)|(?<!\?)\?(?!\?)|(?<!\,)\,(?!\,)|(?<!\;)\;(?!\;)|(?<!\:)\:(?!\:)', r'\g<0> ', text)
    text = re.sub(r'(\.{2,}|\!{2,}|\?{2,})', r'\1 ', text)
    
    return text

df['Split_Review'] = df['Split_Review'].apply(add_space_after_punctuation)

```

```python
df['Split_Review'] = df['Split_Review'].str.replace(r'\s+', ' ', regex=True)
```

```python
df.to_csv("../Data/output/data_for_vader.csv", index=False)
```
